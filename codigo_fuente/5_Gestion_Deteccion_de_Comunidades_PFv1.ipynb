## Cargando la data


```python
# Cargar el grafo desde el archivo GraphML
gestion = nx.read_graphml('gestion.graphml')
```

##Comunidades

```python
from cdlib import algorithms
from cdlib import viz
from cdlib import evaluation
```

## Método Louvain

```python
louvain_com = algorithms.louvain(gestion)
```

```python
print('La cantidad de comunidades es {}'.format(len(louvain_com.communities)))
for c in louvain_com.communities:
  print(c)
```

```python
#mostrar la modularidad general
print(louvain_com.modularity_density())
#mostrar la modularidad de cada comunidad
print(louvain_com.internal_edge_density(summary=False))
```

El resultado muestra la modularidad general de la red, que es 165.68, y la modularidad de cada comunidad individual. Los valores más altos indican una mejor estructura de comunidad.

```python
#calcular los grados de los nodos del grafo y los almacenarlos
d = {}
for x in gestion.degree():
  d[x[0]] = x[1]
print(d)
```

```python
#para cada comunidad
for c in louvain_com.communities:
  #calcular la estación de grado máximo
  max = 0
  s_max = 0
  for s in c: #para cada estación en la comunidad c
    if max < d[s]:
      max = d[s]
      s_max = s
  #imprimir el resultado para la comunidad c
  print(s_max, '\t', max, '\t', c)
```

##Visualización

```python
#vizualizar el grafo según comunidades
plt.figure(figsize=(15, 10))
pos = nx.spring_layout(gestion,seed=30)
viz.plot_network_clusters(gestion, louvain_com, pos, node_size=50)
plt.show()
```

```python
#visualizar el grafo de comunidades
plt.figure(figsize=(15, 10))
viz.plot_community_graph(gestion, louvain_com)
plt.show()
```

```python
# METODO LEIDEN
Leiden_com = algorithms.leiden(gestion)

print('La cantidad de comunidades es {}'.format(len(Leiden_com.communities)))
for c in Leiden_com.communities:
  print(c)

#mostrar la modularidad general
print("")
print("LA MODULARIDAD DE METODO DE LEIDEN ES")
print(Leiden_com.modularity_density())
```

 la visualización del grafo según comunidades te permite obtener una comprensión intuitiva de cómo se estructura la red de usuarios o contenido del Diario Gestión en comunidades distintas, lo que puede proporcionar información valiosa para analizar patrones de interacción, preferencias de los usuarios y la dinámica de la comunidad en línea.

Este gráfico facilita la comprensión visual de la estructura del grafo, identificamos patrones de conexión y relaciones entre nodos, así como visualizar de manera efectiva las comunidades detectadas o cualquier otra información relevante sobre la topología del grafo.

```python
#visualizar el grafo de comunidades
plt.figure(figsize=(15, 10))
viz.plot_community_graph(gestion, Leiden_com)
plt.show()
```

Se pueden ver todas las aristas y nodos del grafo. Las comunidades se resaltan con diferentes colores, lo que permite identificar visualmente las comunidades dentro del grafo.

# Método de Propagación de Etiquetas

```python
# METODO DE Propagación de Etiquetas
label_propagation_com = algorithms.label_propagation(gestion)

# Imprimir el número de comunidades detectadas y las comunidades
print("La cantidad de comunidades es", len(label_propagation_com.communities))
for i, community in enumerate(label_propagation_com.communities):
    print(f"Comunidad {i + 1}: {community}")

# Calcular y mostrar la modularidad general
print("\nLa modularidad del método de Propagación de Etiquetas es:")
print(label_propagation_com.modularity_density())
```

```python
#vizualizar el grafo según comunidades
plt.figure(figsize=(15, 10))
pos = nx.spring_layout(gestion,seed=30)
viz.plot_network_clusters(gestion, label_propagation_com, pos, node_size=50)
plt.show()
```

```python
#visualizar el grafo de comunidades
plt.figure(figsize=(15, 10))
viz.plot_community_graph(gestion, label_propagation_com)
plt.show()
```

```python

#mostrar la modularidad general
print("LA MODULARIDAD DE METODO DE louvain_com ES",louvain_com.modularity_density())
print("LA MODULARIDAD DE METODO DE LEIDEN ES",Leiden_com.modularity_density())
print("La modularidad del método de Propagación de Etiquetas es:",label_propagation_com.modularity_density())
```

**Modularidad del método Louvain:** El valor de la modularidad es aproximadamente 145. Esto indica que la partición del grafo realizada por el algoritmo Louvain tiene una buena calidad, ya que la estructura de la red está bien capturada por las comunidades identificadas.

**Modularidad del método Leiden:** El valor de la modularidad es aproximadamente 157.57. Es un valor bastante alto, lo que sugiere que la partición realizada por el algoritmo de Leiden también es de buena calidad.

**Modularidad del método de Propagación de Etiquetas:** El valor de la modularidad es aproximadamente 168.40. Este es el valor más alto de los tres métodos. Indica que la partición del grafo realizada por el método de Propagación de Etiquetas es la que mejor captura la estructura de la red en comunidades significativas.

La modularidad nos permite evaluar la calidad de las particiones de comunidades en un grafo y nos ayuda a entender cómo bien los métodos de detección de comunidades están identificando estructuras significativas en la red. Un valor más alto de modularidad sugiere una mejor partición y una captura más precisa de la estructura de la red en comunidades distintas.
